```python
# en yüksek paunla şampiyon olan takım
maxpuan=lig['puan_düzeltilmis'].max()
lig[lig['puan_düzeltilmis']==maxpuan]
```

```python
# en düşük puanla şampiyon takım
minpuan=lig['puan_düzeltilmis'].min()
lig[lig['puan_düzeltilmis']==minpuan]
```

```python
# en yüksek paun oratlaması ile şampiyon olan takım
lig['ort_puan']=lig['puan_düzeltilmis']/lig['oyun']

maxortpuan=lig['ort_puan'].max()
lig[lig['ort_puan']==maxortpuan]
```

```python
# en az paun ortalaması ile şampiyon olan takım
minortpuan=lig['ort_puan'].min()
lig[lig['ort_puan']==minortpuan]
```

```python
# Şampiyon Olan Takımlar Maç Başı Ortalama Gol Sayıları
lig['ort_gol']=lig['atilan_gol']/lig['oyun']
```

```python
trump=trump.dropna() #non yapıları dışarı atıyor
```

```python
all=" ".join(trump['tweet'])
```

```python
all_list=all.split(" ")
```

```python
etiketler=[]
for kelime in all_list:
  if len(kelime)>0:
    if kelime[0]=="@":
      etiketler.append(kelime)
```

```python
#ödevvvvvvvvvvvvvv
#etiket_tweet("@hasancagrigungor")
#ilgili tweetler df olarak gelecek
#şuana kadar etiketlediği ve o etikete kaç tweet atılmışsa df olacak
#ulke_tweet("Turkey") denilince o ülke adına kaç tweet atılmış
#11.09.2022 tarihli derse geçtik
```

```python
#şuana kadar etiketlediği ve o etikete kaç tweet atılmışsa df olacak
sozluk={}
for etiket in etiketler:
  sozluk[etiket]=etiketler.count(etiket)
```

```python
tagdf=pd.DataFrame()
```

```python
tagdf['tag']=sozluk.keys()
tagdf['adet']=sozluk.values()
```

```python
tagdf.head(10)
```

```python
#etiketleme sayısına göre sıralamak istiyoruz
tagdf.sort_values(by='adet',ascending=False)
#ascending küçükten büyüğe demek ASC
#descending büyükte küçüğe DESC
```
